```python
def extract_max(logfile,varname,step=48):
    with open(logfile,"r") as fi:
        tempdata = []
        for ln in fi:
            if ln.startswith(" "+varname):
                # notice the white space at the beginning of the line
                tempdata.append(ln.split()[3])
    return np.array(tempdata)[::step]
```

```python
#logfile=datadir+"C48N20_dt1800ns8.log"
logfile=datadir+"C48N20_dt3600ns16.log"

varnames = ["plume0{0}".format(i) for i in range(1,9)]

plt.figure(figsize=(16,4))

plt.subplot(1,2,1)
for varname in varnames[0:4]:
    maxVMR = extract_max(os.environ['HOME']+"/FV3/output/C48/dt1800ns8/C48N20_dt1800ns8.log",varname)
    plt.plot(maxVMR,label=varname)
plt.legend()

plt.subplot(1,2,2)
for varname in varnames[0:4]:
    maxVMR = extract_max(os.environ['HOME']+"/FV3/output/C48/dt225ns1/C48N20_dt225ns1.log",varname,step=48*8)
    plt.plot(maxVMR,label=varname)
plt.legend()
```
